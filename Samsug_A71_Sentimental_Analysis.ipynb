```
import nltk
nltk.download('vader_lexicon')
```

```
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```
data = pd.read_csv('Samsung A71_2.csv')
data.head()
```

```
len(data['Sentiment'].unique())
```

```
data['Sentiment'].value_counts()
```

```
score = []
for i in range(len(data)):
  Review = data['Review Title'][i]
  score.append(sia.polarity_scores(Review))
```

```
data["Score"]= score
data.head()
```

```
compound = []
for i in range (len(data)):
  cm = data['Score'][i]
  compound.append(cm['compound'])
```

```
data['Compound'] = compound
data.head()
```

```
sentiment = []
for i in range(len(data)):
  if data['Compound'][i] >= 0:
     sentiment.append('pos')
  elif data['Compound'][i] < 0:
     sentiment.append('neg')
```

```
data["Predicted Sentiment"] = sentiment
data.head()
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(data['Sentiment'],data['Predicted Sentiment'])
cm
```

```
from sklearn.metrics import accuracy_score
a = accuracy_score(data['Sentiment'],data['Predicted Sentiment'])
a
```
